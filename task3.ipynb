```python
infile = open('CRISPR.targets_single.fa','r')
outfile = open('CRISPR_sgRNAs.fa','w')

urt3 = 'GUUUUAGAGCUAGAAAUAGCAAGUUAAAAUAAGGCUAGUCCGUUAUCAACUUGAAAAAGUGGCACCGAGUCGGUGCUUUU'

for line in infile.readlines():
    if line.startswith('>'):
        outfile.write(line)
    else: 
        line = line.replace('T','U')   #rna has u instead of t
        line = line.strip() + urt3 +'\n'
        outfile.write(line)

infile.close()
outfile.close()

outfile = open('CRISPR_sgRNAs.fa','r')

print(outfile.read())
```
